```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('householdtask3.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Bar chart for total households over the years
plt.figure(figsize=(12, 6))
plt.bar(df['year'], df['tot_hhs'], color='skyblue')
plt.title('Total Households Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Households')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Line chart for ownership proportions over the years
plt.figure(figsize=(12, 6))
plt.plot(df['year'], df['own_prop'], marker='o', linestyle='-', color='blue', label='Ownership Proportion')
plt.plot(df['year'], df['own_wm_prop'], marker='s', linestyle='--', color='green', label='Washing Machine Ownership Proportion')
plt.title('Ownership Proportions Over the Years')
plt.xlabel('Year')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Box plot for income distribution
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['income'], color='lightcoral')
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

```
