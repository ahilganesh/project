```python
#Question 1: What is the distribution of video views?
pit.figure(figsize=(10, 6))
sns.histplot(df['views'], bins=50, kde=True, color='skyblue')
pit.title('Distribution of Video Views')
pit.xlabel('Views')
pit.ylabel('Frequency')
pit.show()
```

```python
#Question 2: What is the relationship between likes and dislikes?
pit.figure(figsize=(10, 6))
sns.scatterplot(x='likes', y='dislikes', data=df, alpha=0.5)
pit.title('Relationship Between Likes and Dislikes')
pit.xlabel('Likes')
pit.ylabel('Dislikes')
pit.show()
```

```python
#Question 3: How do comment counts vary with views?
pit.figure(figsize=(10, 6))
sns.scatterplot(x='views', y='comment_count', data=df, alpha=0.5, color='green')
pit.title('Views vs. Comment Count')
pit.xlabel('Views')
pit.ylabel('Comment Count')
pit.show()
```

```python
#Identify Correlations
correlation_matrix = df[['views', 'likes', 'dislikes', 'comment_count']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
pit.title('Correlation Matrix')
pit.show()
```
